# Best Neighborhood Based On Speeding

```python
import pandas as pd
import numpy as np
%matplotlib inline
import geopandas as gd
```

```python
#import data
speeding = pd.read_csv("data.csv", index_col = "neighborhood", parse_dates = True)
#making sure data has been imported correctly
speeding.head(60)
```

```python
#created some graphs to get an idea of how I want to visually represent the data
max_speed = pd.DataFrame(speeding["max_speed"])
mS = max_speed.sort_values("max_speed", ascending = False)
mS.head(25).plot.bar()

speed_limit = pd.DataFrame(speeding["speed_limit"])
sL = speed_limit.sort_values("speed_limit", ascending = False)
sL.head(25).plot.bar()

percent_over_limit = pd.DataFrame(speeding["percent_over_limit"])
pOL = percent_over_limit.sort_values("percent_over_limit", ascending = False)
pOL.head(25).plot.bar()
```

```python
#remove unneeded columns
speeding["amount_over_limit"] = speeding["max_speed"] - speeding["speed_limit"]
speeding = speeding.drop(columns = ["id", "device_id", "record_oid", "count_start_date", "count_end_date",
                                   "average_daily_car_traffic", "average_daily_bike_traffic", "counter_number",
                                   "counter_type", "speed85_percent", "speed95_percent", "longitude", "latitude", 
                                   "council_district", "ward", "tract", "public_works_division", "pli_division", 
                                   "police_zone", "fire_zone"])
#drop rows with no data
speeding = speeding.dropna()
speeding.head(60)
```

```python
#create graph of average mph over the speed limit cars are driving in neighborhoods
amount_over_limit = pd.DataFrame(speeding["amount_over_limit"])
aOL = amount_over_limit.sort_values("amount_over_limit", ascending = True)
aOL.head(90).plot.bar(figsize = (15, 10))
```

```python
#create map
neighborhoods = gd.read_file("https://data.wprdc.org/dataset/e672f13d-71c4-4a66-8f38-710e75ed80a4/resource/c5a93a8e-03d7-4eb3-91a8-c6b7db0fa261/download/dbd133a206c")
speed_map = neighborhoods.merge(speeding, how = "left", left_on = "hood", right_on = "neighborhood")
speed_map.plot(column = "amount_over_limit", cmap = "inferno", edgecolor = "pink", legend = True, 
               legend_kwds = {"label" : "amount_over_limit"}, figsize = (15, 10))
```

According to the map, the brighter the color, the higher over the speed limit cars tend to travel in a particular neighborhood. Therefore, those neighborhoods colored dark purple and black are the safest neighborhoods based on speeding tendencies. 

According to the data, Spring Hill-City View is the best and safest neighborhood with no cars speeding. Brookline is the most dangerous neighborhood with cars driving 78 mph over the speed limit, making it the worst. 

```python
parks = pd.read_csv("parks.csv")
parks = parks.dropna()

import csv

with open ("parks.csv") as f:
    parks = [row for row in csv.reader(f)]
    
# goes through the data and counts how many pieces of art are in each neighborhood
park_counter = dict()

for row in parks[2:]:
    neighborhood = row[5]
    
    if neighborhood not in park_counter:
        park_counter[neighborhood] = 1
    else:
        park_counter[neighborhood] = park_counter[neighborhood] + 1
park_counter  
```

```python
#split dict into neighborhoods and number of pieces of art
neighborhoods = []
number_of_parks = []
items = park_counter.items()
for item in items:
    neighborhoods.append(item[0]), number_of_parks.append(item[1])
    
#create dataframe using my two dictionaries
park_df = pd.DataFrame({"neighborhood":neighborhoods,"number of parks":number_of_parks})

#sorted the dataframe from greatest to least
park_df.sort_values(by = "number of parks", ascending = False)
```
